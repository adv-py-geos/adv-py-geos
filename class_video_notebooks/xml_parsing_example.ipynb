```python
import xml.etree.ElementTree as ET

tree = ET.parse('rows.xml') 
tree
```

```python
root = tree.getroot()
root
```

```python
root[0][0].find('college_1yr_more_male')
```

```python
for child in root[0]:
    print(child.find('fips_id').text)
```

```python
for child in root.iter('high_school_more_female'):
    print(child.text)
```

`response` is the 'root' element. You will see it is the first element in the file above, in angle brackets.
