```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

- I just read csv files with 'skiprows=8' to use the given columns.

```python
#Import csv files in python as DataFrame
df_cdt = pd.read_csv('conductivity.csv',skiprows=8)
df_sst = pd.read_csv('water_temp.csv',skiprows=8)
df_chl = pd.read_csv('rel_chlor.csv',skiprows=8)
df_cdt['DateTime']=pd.to_datetime(df_cdt['DateTime'], infer_datetime_format=True)
df_sst['DateTime']=pd.to_datetime(df_sst['DateTime'], infer_datetime_format=True)
df_chl['DateTime']=pd.to_datetime(df_chl['DateTime'], infer_datetime_format=True)
df_cdt
```

- Extract and rename the columns

```python
data_cdt={'Date/Time': df_cdt['DateTime'],
               'Value[mS/cm]': df_cdt[' Measured Value']}
cdt= pd.DataFrame(data_cdt,columns = ['Date/Time','Value[mS/cm]'])
cdt['Value[mS/cm]']=cdt['Value[mS/cm]']/1000

data_sst={'Date/Time': df_sst['DateTime'],
               'Value[℃]': df_sst[' Measured Value']}            
sst= pd.DataFrame(data_sst,columns = ['Date/Time','Value[℃]'])

data_chl={'Date/Time': df_chl['DateTime'],
               'Value[RFU]': df_chl[' Measured Value']}            
chl= pd.DataFrame(data_chl,columns = ['Date/Time','Value[RFU]'])
```

```python
cdt
```

```python
# Main Plot
plt.figure(figsize=(15,9))
plt.subplots_adjust(hspace = .4)

# SST
plt.subplot(311)
plt.plot(sst['Date/Time'], sst['Value[℃]'], 'r--')
plt.title("Sea Surface Temperature Plot")
plt.xlabel('Date')
plt.ylabel('[℃]')
plt.grid()


# Conductivity
plt.subplot(312)
plt.plot(cdt['Date/Time'], cdt['Value[mS/cm]'], 'b--')
plt.title("Conductivity Plot")
plt.xlabel('Date')
plt.ylabel('[mS/cm]')
plt.grid()

# Chlorophyll
plt.subplot(313)
plt.plot(chl['Date/Time'], chl['Value[RFU]'], 'g--')
plt.title("Chlorophyll Plot")
plt.xlabel('Date')
plt.ylabel('[RFU]')
plt.grid()
```
